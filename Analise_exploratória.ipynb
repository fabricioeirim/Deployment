```python
df=pd.read_excel("/content/AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Total de receita
df["Valor Venda"].sum()
```

```python
#Criando a coluna de custo total
df["Custo"]=df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
round(df["Custo"].sum(),2)
```

```python
df["Lucro"]=df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
#Lucor total
round(df["Lucro"].sum(),2)
```

```python
df["Tempo de Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
df["Tempo de Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
df["Tempo de Envio"].dtypes 
```

```python
df.groupby("Marca")["Tempo de Envio"].mean()
```

```python
df.isnull().sum()
```

```python
#Agrupando por Data da venda e marca e trazendo o lucro
from pandas.core.groupby import groupby
from pandas.core.groupby.groupby import group_selection_context
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Formatando a vizualização dos numeros.
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Trazer a iformação do agrupamento em formato de colunas
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Produto")
plt.legend
plt.ylabel("Quantidade");
```

```python
df.groupby([df["Data Venda"].dt.year])["Lucro"].sum().plot.bar(title="Lucro por Ano")
plt.xlabel("Ano")
plt.legend
plt.ylabel("Lucro");
```

```python
df.groupby([df["Data Venda"].dt.year])["Lucro"].sum()
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.legend
plt.ylabel("Lucro");
```

```python
df_2009.groupby(df["Marca"])["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.legend
plt.ylabel("Lucro");
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby(df["Classe"])["Lucro"].sum().plot.pie(title="Lucro x Classe")
plt.xlabel("Classe")
plt.legend
plt.ylabel("Lucro")
```
