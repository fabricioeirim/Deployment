```python
#Realizando a iortação das bibliotecas e funções de bibliotecas
#pandas realiza tratamento do dtaset
#numpy , biblioteca para computação numerica
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
#Extraindo a base de dados.
df = pd.read_csv('/content/train.csv')
```

```python
#Lendo as 10 primeiras linhas para verificar e validar os campos do dataset.
df.head()
```

```python
#Excluindo colunas que não serão utilizadas na analise explratória.
df=df.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
df.head()
```

```python
#Definindo uma chave primária no dataset, utilizanod a coluna PassengerId
df=df.set_index(['PassengerId'])
df=df.rename(columns = {'Survived': 'target'}, inplace = False)
```

```python
df.head()
```

```python
#função de analise descritiva simples
df.describe()
```

```python
#Resumo analise descritiva
df.describe(include=['O'])
```

```python
#Ajustando classificação para algoritimo de treino
df['SEX_f'] = np.where(df['Sex'] == 'female',1,0)
df['Pclass_1'] = np.where(df['Pclass']== 1,1,0)
df['Pclass_2'] = np.where(df['Pclass']== 2,1,0)
df['Pclass_3'] = np.where(df['Pclass']== 3,1,0)
```

```python
#removendo colunas utilizadas para reclassificação
df=df.drop(['Pclass', 'Sex'], axis=1)
```

```python
df.head()
```

```python
#verificando arquivos nulos
df.isnull().sum()
```

```python
#Removendo valores nulos e subistituindo por 0
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
#estruturando a base para imputar no algoritimo de treino, 30% para treino do algoritimo,
# e 70% para teste do algoritimo.
from pandas.core.common import random_state
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis =1),
df["target"],
test_size = 0.3,
random_state = 1234)
[{'treino': x_train.shape}, {'teste':x_test.shape}]
```

```python
#Algoritimo Randon Forest para Machine Leraning (Análise Preditiva)
rndforest = RandomForestClassifier(n_estimators=1000,
                                   criterion='gini',
                                   max_depth=5)

rndforest.fit(x_train, y_train)
```

```python
#Criando as variaveis, probalidade e classificação , baseadas no algoritimo de treino e teste.
probabilidade = rndforest.predict_proba(df.drop('target', axis=1))[:,1]
classificacao = rndforest.predict(df.drop('target', axis=1))
```

```python
df['probabilidade'] = probabilidade *100
df['classificacao'] = classificacao
```
