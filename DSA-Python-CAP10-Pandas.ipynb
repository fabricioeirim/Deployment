```python
#Extraímos a moda da coluna Quantidade
moda = dsa_df['Quantidade'].value_counts().index[0]
```

## A moda estatistica é uma medida de tendencia central que representa o valor mais frequente em um conjunto de dados. A moda é especialmente útil quando queremos saber qual é o valor mais comum ou popular em um conjunto de dados, seja em uma distribuição unimodal (com apenas uma medida) ou uma distrobuição bimodal (com duas modas). No entando a moda pode não ser uma medida tão representativa quanto outras medidas de tendencia central como a média ou a mediana, especialmente em distribuições assimétricas ou quando a valores extremos. Por essa razão é importante analisar diferentes medidas de tendencia central e usar a que melhor se adequa aos seus objetivos na análise estatistica.

```python
print(moda)
```

```python
#E por fim preenchemos os valores na com a moda
dsa_df['Quantidade'].fillna(value = moda, inplace = True)
```

## Operadores lógicos para manipuação de dados com o Pandas.


##### Operador AND

```python
#Filtrando as vendas que ocorrem no segmento home-office na região sul
# O [] indica o slice e o () é onde colocamos a expressão, essa é a regra para aplcarmos expressões dentro de slices, usando 
# o dataframe.coluna dentro de parenteses e todo o slice dentro de colchetes.
#O & representa o operador AND
dsa_df4 = dsa_df[(dsa_df.Segmento == 'Home Office') & (dsa_df.Regiao=='South')].head()
```

```python
dsa_df4
```

##### Operador OR

```python
#Filtrando as vendas que ocorrem no segmento home-office ou na região sul
# O [] indica o slice e o () é onde colocamos a expressão, essa é a regra para aplcarmos expressões dentro de slices, usando 
# o dataframe.coluna dentro de parenteses e todo o slice dentro de colchetes.
#O | representa o operador OR
dsa_df5 = dsa_df[(dsa_df.Segmento == 'Home Office') | (dsa_df.Regiao=='South')].tail()
```

```python
dsa_df5
```

##### Operador Negação

```python
# != significa "Diferente em Python"
#Método Sample retorna amostras aleatórias
dsa_df6 = dsa_df[(dsa_df.Categoria != 'Home Office') & (dsa_df.Regiao != 'South')].sample(5)
```

```python
dsa_df6
```

### Agrupamento de dados em DataFrames com o GroupBy

```python
#Aplicando o grupBy - Seria o mesmo que fazer uma tabela Dinamuica no Excel 
#O agrupamento precisa ser por colunas não numéricas
#O groupBy precisa de uma função matemática de uma coluna que não aparece no agrupamento, ou seja, uma coluna numerica.
dsa_df[['Segmento','Regiao','Valor_Venda']].groupby(['Segmento','Regiao']).mean()
```

```python
dsa_df[['Regiao','Categoria','Quantidade']].groupby(['Regiao','Categoria']).count()
```

```python
dsa_df.dtypes
```

#### Agragação multipla com o GroupBY - Vamos explorar mais a função groupby() pois temos diversas opções de sumarização
#### de dados de forma simples. No exemplo abaixo uniremos a função groupBy com a função agg que realiza agragação multipla.

```python
#Aplicanod groupby + agg
dsa_df[['Segmento','Regiao','Valor_Venda']].groupby(['Segmento','Regiao']).agg(['mean','std','count'])
```

```python
#Filtrando a partir d strings - Utilizando o método startswith -> filtra pelo começo de uma string, ex: startswith 'Con'
dsa_df[dsa_df.Segmento.str.startswith('Con')]
```

```python
dsa_df.Segmento.value_counts()
```

```python
#Filtrando a partir d strings - Utilizando o método endswith -> filtra pelo final de uma string, ex: endswith 'mer'
dsa_df[dsa_df.Segmento.str.endswith('mer')].head()
```
